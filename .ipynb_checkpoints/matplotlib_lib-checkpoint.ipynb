```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r =10
theta = np.linspace(0,2*np.pi,200)
```

```python
x=r*np.cos(theta)
y=r*np.sin(theta)
```

```python
plt.figure(figsize = (4,4))#given size of circle
plt.plot(x,y)
```
